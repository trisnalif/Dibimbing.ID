```python
# Info pokemon dengan Total paling tinggi
pokemon_data['Total'] = pokemon_data[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].sum(axis=1)

# Mengurutkan dan Memilih 5 PokÃ©mon dengan Total Terbesar
top_10_pokemon = pokemon_data.nlargest(10, 'Total')

# Menampilkan Informasi
print("10 PokÃ©mon dengan Total Stat Terbesar:")
print(top_10_pokemon)
```

```python
# Plot total stat Pokemon
plt.figure(figsize=(10, 6))
bars = plt.bar(top_10_pokemon['Name'], top_10_pokemon['Total'], color=['#ff9999', '#66b3ff', '#99ff99'])

# Menambahkan label total stat di atas setiap bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval}', va='bottom', ha='center')  # va: vertical alignment, ha: horizontal alignment

plt.xlabel('PokÃ©mon')
plt.ylabel('Total Stat')
plt.title('10 PokÃ©mon dengan Total Stat Terbesar')


plt.xticks(rotation=45, ha='right')

plt.tight_layout()  # Menyesuaikan tata letak agar label tidak terpotong
plt.show()
```

```python
# Nama PokÃ©mon yang dicari
pokemon_name = 'LatiasMega Latias'  # Ganti dengan nama PokÃ©mon yang diinginkan

# Menampilkan informasi tentang PokÃ©mon dengan nama tertentu
pokemon_info = pokemon_data[pokemon_data['Name'] == pokemon_name]

# Menampilkan hasil dalam format tabel
print(pokemon_info.to_string(index=False))
```

```python
# Distribusi tipe Pokemon
pokemon_data['num_types'] = pokemon_data.apply(lambda row: 2 if pd.notna(row['Type 2']) else 1, axis=1)

# Menghitung jumlah PokÃ©mon dengan satu tipe dan dua tipe
type_counts = pokemon_data['num_types'].value_counts()

# Menghitung Persentase
type_percentages = type_counts / type_counts.sum() * 100

# Memplot Pie Chart
labels = ['Tipe 1', 'Tipe 2']
colors = ['#ffcc99','#99ccff']
plt.figure(figsize=(8, 6))
wedges, texts, autotexts = plt.pie(type_percentages, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors )
plt.legend(wedges, labels, title="Jumlah Tipe", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('Persentase PokÃ©mon Berdasarkan Jumlah Tipe')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()
```

```python
# Distribusi Generasi Pokemon
# Menghitung Distribusi Generasi
generation_counts = pokemon_data['Generation'].value_counts().sort_index()

# Menghitung Persentase
generation_percentages = generation_counts / generation_counts.sum() * 100

# Memplot Bar Chart
labels = [f'Generasi {gen}' for gen in generation_percentages.index]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6','#ff6666']  # Warna berbeda untuk setiap generasi

plt.figure(figsize=(12, 8))
bars = plt.bar(labels, generation_percentages, color=colors)

# Menambahkan label persentase di atas setiap bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.1f}%', va='bottom')  # va: vertical alignment

plt.xlabel('Generasi')
plt.ylabel('Persentase')
plt.title('Persentase PokÃ©mon Berdasarkan Generasi')
plt.show()
```

```python
#Persentase pokemon legendary dan non-legendary

# Menghitung Distribusi PokÃ©mon Legendary dan Non-Legendary
legendary_counts = pokemon_data['Legendary'].value_counts()

# Menghitung Persentase
legendary_percentages = legendary_counts / legendary_counts.sum() * 100

# Memplot Pie Chart
labels = ['Non-Legendary', 'Legendary']
colors = ['#ff9999', '#66b3ff']  # Warna berbeda untuk setiap kategori

plt.figure(figsize=(8, 6))
wedges, texts, autotexts = plt.pie(legendary_percentages, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, wedgeprops=dict(width=0.3))

# Menambahkan legenda
plt.legend(wedges, labels, title="Kategori", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Persentase PokÃ©mon Legendary dan Non-Legendary')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()
```

```python
# sorting pokemon by Type
type1_counts = pokemon_data['Type 1'].value_counts()
type2_counts = pokemon_data['Type 2'].value_counts()
```

```python
total_counts = type1_counts.add(type2_counts, fill_value=0).sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=total_counts.values, y=total_counts.index, palette='viridis')
plt.xlabel('Jumlah PokÃ©mon')
plt.ylabel('Tipe PokÃ©mon')
plt.title('Distribusi Tipe PokÃ©mon')
plt.show()
```

```python
# Sorting Legendary dan non-Legendary pokemon
legendary_pokemon = pokemon_data[pokemon_data['Legendary'] == True]
non_legendary_pokemon = pokemon_data[pokemon_data['Legendary'] == False]
```

```python
legendary_type1_counts = legendary_pokemon['Type 1'].value_counts()
legendary_type2_counts = legendary_pokemon['Type 2'].value_counts()
non_legendary_type1_counts = non_legendary_pokemon['Type 1'].value_counts()
non_legendary_type2_counts = non_legendary_pokemon['Type 2'].value_counts()
```

```python
legendary_total_counts = legendary_type1_counts.add(legendary_type2_counts, fill_value=0).sort_values(ascending=False)
non_legendary_total_counts = non_legendary_type1_counts.add(non_legendary_type2_counts, fill_value=0).sort_values(ascending=False)
```

```python
# Ploting Legendary dan non-Legendary pokemon
plt.figure(figsize=(14, 8))
```

```python
# Plot untuk PokÃ©mon Legendaris
plt.subplot(1, 2, 1)
sns.barplot(x=legendary_total_counts.values, y=legendary_total_counts.index, palette='viridis')
plt.xlabel('Jumlah PokÃ©mon')
plt.ylabel('Tipe PokÃ©mon')
plt.title('Distribusi Tipe PokÃ©mon Legendaris')

# Plot untuk PokÃ©mon Non-Legendaris
plt.subplot(1, 2, 2)
sns.barplot(x=non_legendary_total_counts.values, y=non_legendary_total_counts.index, palette='viridis')
plt.xlabel('Jumlah PokÃ©mon')
plt.ylabel('Tipe PokÃ©mon')
plt.title('Distribusi Tipe PokÃ©mon Non-Legendaris')

plt.tight_layout()
plt.show()
```
